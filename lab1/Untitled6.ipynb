```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.loadtxt('X_out.csv')
x1 = data1[:, 0]
y1 = data1[:, 1]

plt.plot(x1, y1, label='X(t)')

data2 = np.loadtxt('V_out.csv')
x2 = data2[:, 0]
y2 = data2[:, 1]

plt.plot(x2, y2, label='V(t)')

plt.xlabel('t')
plt.ylabel('x,v')
plt.title('График функции X(t), V(t)')

plt.legend()

plt.show()
```
